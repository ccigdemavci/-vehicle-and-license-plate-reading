```python
from PIL import Image
import os

def resize_with_padding(image, target_size=(640, 640), fill_color=(114, 114, 114)):
    original_width, original_height = image.size
    ratio = min(target_size[0] / original_width, target_size[1] / original_height)
    new_size = (int(original_width * ratio), int(original_height * ratio))
    image = image.resize(new_size, Image.BILINEAR)

    new_image = Image.new("RGB", target_size, fill_color)
    paste_position = ((target_size[0] - new_size[0]) // 2, (target_size[1] - new_size[1]) // 2)
    new_image.paste(image, paste_position)

    return new_image

def process_images(input_folder, output_folder, target_size=(640, 640)):
    os.makedirs(output_folder, exist_ok=True)
    for filename in os.listdir(input_folder):
        if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            img_path = os.path.join(input_folder, filename)
            img = Image.open(img_path).convert("RGB")
            padded_img = resize_with_padding(img, target_size)
            padded_img.save(os.path.join(output_folder, filename))

# Kullanım:
process_images("kamyon", "dataset/kamyon", target_size=(640, 640))
process_images("araba", "dataset/araba", target_size=(640, 640))
```

```python
import os
import cv2
import numpy as np

def augment_and_save(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if not filename.lower().endswith(('.jpg', '.jpeg', '.png')):
            continue

        path = os.path.join(input_dir, filename)
        img = cv2.imread(path)
        if img is None:
            continue

        base_name = os.path.splitext(filename)[0]
        h, w = img.shape[:2]

        # 1. Orijinal
        cv2.imwrite(os.path.join(output_dir, f"{base_name}_orig.jpg"), img)

        # 2. Hafif döndürme (örneğin 10 derece)
        angle = 10
        M = cv2.getRotationMatrix2D((w/2, h/2), angle, 1.0)
        rotated = cv2.warpAffine(img, M, (w, h))
        cv2.imwrite(os.path.join(output_dir, f"{base_name}_rot.jpg"), rotated)

        # 3. Parlaklık artırma (gündüz etkisi)
        bright = cv2.convertScaleAbs(img, alpha=1.2, beta=30)
        cv2.imwrite(os.path.join(output_dir, f"{base_name}_bright.jpg"), bright)

print("🔁 Kamyonlar çoğaltılıyor...")
augment_and_save("dataset/kamyon", "veri_aug/kamyon_aug")

print("🔁 Arabalar çoğaltılıyor...")
augment_and_save("dataset/araba", "veri_aug/araba_aug")

print("✅ Veri çoğaltma tamamlandı!")
```

```python
import os
import time
from roboflow import Roboflow
import cv2

class_name_to_id = {
    "car": 0,
    "truck": 1,
    "plate": 2
}

rf = Roboflow(api_key="gpVGykH7opfhZEoE84Uj")
project = rf.workspace("goruvize").project("araba-kamyon")
model = project.version(1).model

input_dirs = [
    "/Users/cigdemavci/Desktop/bilgisayai_goru_vize/veri_aug/araba_aug/train"
]

for input_dir in input_dirs:
    output_label_dir = os.path.join(input_dir.replace("veri_aug", "yolo_dataset/labels"))
    os.makedirs(output_label_dir, exist_ok=True)

    for i, filename in enumerate(os.listdir(input_dir)):
        if not filename.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        image_path = os.path.join(input_dir, filename)
        label_filename = os.path.splitext(filename)[0] + ".txt"
        label_path = os.path.join(output_label_dir, label_filename)

        if os.path.exists(label_path):
            continue  # Zaten etiketlenmiş

        try:
            print(f"🔍 Etiketleniyor: {filename}")
            time.sleep(1)
            prediction = model.predict(image_path, confidence=40, overlap=30).json()
            image = cv2.imread(image_path)
            h, w, _ = image.shape

            with open(label_path, "w") as f:
                for obj in prediction["predictions"]:
                    class_name = obj["class"]
                    class_id = class_name_to_id.get(class_name, -1)
                    if class_id == -1:
                        continue
                    x_center = obj["x"] / w
                    y_center = obj["y"] / h
                    width = obj["width"] / w
                    height = obj["height"] / h
                    f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
        except Exception as e:
            print(f"❌ Hata: {filename} → {str(e)}")

print("✅ Etiketleme tamamlandı!")
```

```python
import os
import random
import shutil
from roboflow import Roboflow

# === ROBOT AYARLARI ===
API_KEY = "gpVGykH7opfhZEoE84Uj"  # <- kendi Roboflow API anahtarını gir
WORKSPACE = "goruvize"  # <- roboflow workspace adı (genelde username ile aynı)
PROJECT = "araba-kamyon"      # <- roboflow proje adı (URL'den alabilirsin)
VERSION = "1"                 # <- model versiyon numarası

# === Sınıf isimleri ve ID'leri ===
class_map = {
    "araba": 0,
    "kamyon": 1,
    "plaka": 2
}

# === Klasör yolları ===
input_dir = os.path.expanduser("~/Desktop/bilgisayai_goru_vize/veri_aug")
class_folders = ["kamyon_aug", "araba_aug"]  # kaynak alt klasörler
output_base = os.path.expanduser("~/Desktop/bilgisayai_goru_vize/dataset_yolo_ready")

# === Roboflow modelini başlat ===
rf = Roboflow(api_key=API_KEY)
project = rf.workspace(WORKSPACE).project(PROJECT)
model = project.version(VERSION).model

# === Çıktı klasörlerini oluştur ===
splits = ["train", "val"]
for split in splits:
    os.makedirs(os.path.join(output_base, "images", split), exist_ok=True)
    os.makedirs(os.path.join(output_base, "labels", split), exist_ok=True)

# === Görselleri topla ===
image_paths = []
for class_folder in class_folders:
    folder_path = os.path.join(input_dir, class_folder)
    for file in os.listdir(folder_path):
        if file.lower().endswith((".jpg", ".jpeg", ".png")):
            image_paths.append(os.path.join(folder_path, file))

# === Karıştır ve böl ===
random.shuffle(image_paths)
split_idx = int(len(image_paths) * 0.8)
train_images = image_paths[:split_idx]
val_images = image_paths[split_idx:]

# === Tahmin yap + etiketle + kopyala ===
def process_and_save(images, split):
    for image_path in images:
        filename = os.path.basename(image_path)
        name, _ = os.path.splitext(filename)

        try:
            prediction = model.predict(image_path, confidence=40, overlap=30).json()
        except Exception as e:
            print(f"❌ Hata oluştu: {image_path} - {e}")
            continue

        # Görsel boyutlarını al
        try:
            image_width = float(prediction["image"]["width"])
            image_height = float(prediction["image"]["height"])
        except KeyError:
            print(f"❌ Görsel boyutları alınamadı: {image_path}")
            continue

        # Etiket dosyasını yaz
        label_path = os.path.join(output_base, "labels", split, f"{name}.txt")
        with open(label_path, "w") as f:
            for pred in prediction["predictions"]:
                class_name = pred["class"]
                class_id = class_map.get(class_name, -1)

                if class_id == -1:
                    print(f"⚠️ Bilinmeyen sınıf: {class_name} — {image_path}")
                    continue

                try:
                    x_center = float(pred["x"]) / image_width
                    y_center = float(pred["y"]) / image_height
                    width = float(pred["width"]) / image_width
                    height = float(pred["height"]) / image_height
                except (ValueError, KeyError, TypeError) as e:
                    print(f"❌ Sayısal dönüşüm hatası: {e} — {image_path}")
                    continue

                f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")

        # Görseli kopyala
        target_img_path = os.path.join(output_base, "images", split, filename)
        shutil.copy(image_path, target_img_path)
        print(f"✅ Etiketlendi ve kopyalandı: {filename} → {split}")

# === İşlemleri uygula ===
process_and_save(train_images, "train")
process_and_save(val_images, "val")

print("\n🎉 İşlem tamamlandı: Roboflow etiketleme + YOLO formatında train/val ayrımı")
```
